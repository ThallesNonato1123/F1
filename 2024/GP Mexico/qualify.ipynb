```python
import pandas as pd
import sys
sys.path.append("../..")
from F1Event import F1Event
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
```

```python
saiTel = pd.read_csv("sainzTelemetria.csv")
verTel = pd.read_csv("verTelemetria.csv")
```

```python
saiTel.columns
```

```python
saiTel[["Speed", "Distance", "Throttle","Brake", "DRS", "RPM"]].to_csv("sainzTelemetria.csv", index=False, encoding='utf-8')
```

```python
verTel[["Speed", "Distance", "Throttle","Brake", "DRS", "RPM"]].to_csv("verTelemetria.csv", index=False, encoding='utf-8')
```

```python
fig = make_subplots(rows=3, cols=1, shared_xaxes=True)
```

### Tratamento de Dados

```python
saiTel['Brake'] = np.where(saiTel['Brake'] == True, 1,0)
verTel['Brake'] = np.where(verTel['Brake'] == True, 1,0)
```

### Gr√°ficos de linha

```python
fig.add_trace(go.Scatter(x=saiTel['Distance'], y=saiTel['Speed'], name='Sainz', line=dict(color='red')), row=1, col=1)
fig.add_trace(go.Scatter(x=verTel['Distance'], y=verTel['Speed'], name='Verstappen', line=dict(color='blue')), row=1, col=1)

fig.add_trace(go.Scatter(x=saiTel['Distance'], y=saiTel['Throttle'],  name='Sainz', line=dict(color='red'),  showlegend=False ), row=2, col=1)
fig.add_trace(go.Scatter(x=verTel['Distance'], y=verTel['Throttle'],  name='Verstappen', line=dict(color='blue'),  showlegend=False ), row=2, col=1)

fig.add_trace(go.Scatter(x=saiTel['Distance'], y=saiTel['Brake'],  name='Sainz', line=dict(color='red'),  showlegend=False ), row=3, col=1)
fig.add_trace(go.Scatter(x=verTel['Distance'], y=verTel['Brake'],  name='Verstappen', line=dict(color='blue'),  showlegend=False ), row=3, col=1)



fig.update_yaxes(title_text="Velocidade", row=1, col=1)
fig.update_yaxes(title_text="Acelerador", row=2, col=1)
fig.update_yaxes(title_text="Freio", row=3, col=1)


fig.update_layout(
    title='Capacita√ß√£o Icarus - Gr√°fico de Linha',
)
```

### Histogramas

```python
fig = go.Figure(data=[go.Histogram(
    x=verTel["Speed"],
    nbinsx=5,  
    marker=dict(color='blue'),  
)])

fig.update_layout(
    title='Histograma Velocidade - Capacita√ß√£o Icarus',
    xaxis_title='Valor',
    yaxis_title='Frequ√™ncia',
    bargap=0.1
)
```

```python
fig = go.Figure(data=[go.Scatter(
    x=saiTel["Speed"],
    y = saiTel["RPM"], 
    mode = 'markers',
    marker=dict(color='red'),  
)])

fig.update_layout(
    title='Capacita√ß√£o Icarus - Scatterplot',
)

fig.show()
```
